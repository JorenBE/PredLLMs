```python
from fastcore.xtras import load_pickle
import pandas as pd 
import os 
from glob import glob
from pathlib import Path
from scipy.stats import sem
```

```python
all_res = glob('out/*/*.pkl')
```

```python
compiled_res = []

for res in all_res:
    res = load_pickle(res)
    summary = {
        'size': res['size'],
        'representation': res['representation'],
        'accuracy': res['accuracy'],
        'f1_macro': res['f1_macro'],
        'f1_micro': res['f1_micro'],
        'kappa': res['kappa'],
        'num_epochs': res['num_epochs'],
    }
    compiled_res.append(summary)
```

```python
compiled_res = pd.DataFrame(compiled_res)
```

```python
grouped_res =compiled_res.groupby(['num_epochs', 'representation', 'size']).agg(['mean', 'sem'])
```
